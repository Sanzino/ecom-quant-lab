<h2 style="border-bottom:2px solid #555; padding-bottom:6px">🧩 Konsept 1 — Problemstillingen</h2>
<p>De fleste e-commerce gründere fokuserer på <b>feil metricer</b>:</p>
<ul>
  <li>❌ "Jeg fikk 10 salg i dag!" — men var de lønnsomme?</li>
  <li>❌ "ROAS er 2.5!" — men hva er break-even ROAS?</li>
  <li>❌ "CPA er 150 kr" — men hva er max CPA før du taper penger?</li>
  <li>❌ "Jeg brukte 5000 kr på ads" — men hva er faktisk profit etter alle kostnader?</li>
</ul>

<p><b>Sannheten:</b> Du kan ha høy omsetning, "bra" ROAS, og mange salg — og <i>fortsatt tape penger</i>.</p>

<p style="background:rgba(255,99,71,0.1); padding:10px; border-left:4px solid rgba(255,99,71,0.5); border-radius:3px">
  ⚠️ <b>Kritisk innsikt:</b> Profit kommer ikke fra salgsvolum — det kommer fra <b>margin multiplisert med volum</b>. Hvis marginen er negativ eller for lav, dreper volum deg raskere.
</p>

<p><b>Eksempel fra virkeligheten:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Produkt selges for: 500 kr
Produktkostnad: 200 kr
Frakt: 80 kr
Payment fees (Stripe): 17 kr
👉 Net profit FØR ads: 203 kr

Du bruker 250 kr i ads per salg (CPA = 250 kr)
👉 Faktisk profit ETTER ads: -47 kr

Du taper 47 kr per salg — men føler du "vokser" fordi du får salg.
</pre>

<p>Dette er <b>ikke teoretisk</b> — dette skjer hver dag. Folk scaler seg til konkurs.</p>
<hr>

<h2 style="border-bottom:2px solid #555; padding-bottom:6px">🧩 Konsept 2 — Hva er Net Profit (før ads)?</h2>
<p><b>Net profit</b> er pengene som <i>faktisk</i> blir igjen etter at produktet er solgt — men FØR du har betalt for annonseringen.</p>

<p><b>Formelen:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Net Profit = Sale Price - Product Cost - Shipping - Payment Fees
</pre>

<p><b>Komponentene:</b></p>
<ul>
  <li><b>Sale Price:</b> Hva kunden betaler (inkl. MVA hvis du opererer B2C i Norge)</li>
  <li><b>Product Cost:</b> Hva DU betalte for produktet (COGS - Cost of Goods Sold)</li>
  <li><b>Shipping Cost:</b> Faktisk fraktkostnad (ikke hva kunden betaler, men hva DU betaler)</li>
  <li><b>Payment Fees:</b> Stripe, Klarna, Vipps tar 2-3% + fast fee per transaksjon</li>
</ul>

<p style="background:rgba(100,149,237,0.1); padding:10px; border-left:4px solid rgba(100,149,237,0.5); border-radius:3px">
  💡 <b>Viktig:</b> Dette er IKKE profit etter ads. Dette er "rommet" du har å jobbe med. Hvis net profit er 200 kr, kan du max bruke 200 kr på ads per salg.
</p>

<p><b>Eksempel:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Sale Price: 799 kr
Product Cost: 250 kr
Shipping: 65 kr
Payment Fee (2.9% + 3 kr): ~26 kr

Net Profit = 799 - 250 - 65 - 26 = 458 kr

Dette betyr: Du kan bruke opptil 458 kr på ads og fortsatt gå i null.
</pre>
<hr>

<h2 style="border-bottom:2px solid #555; padding-bottom:6px">🧩 Konsept 3 — Margin %</h2>
<p><b>Margin</b> forteller deg hvor mye av hver krone som faktisk er profit.</p>

<p><b>Formelen:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Margin % = (Net Profit / Sale Price) × 100
</pre>

<p><b>Hva betyr det?</b></p>
<ul>
  <li><b>50% margin:</b> Halvparten av salget er profit (ekstremt bra)</li>
  <li><b>30% margin:</b> 30 øre av hver krone er profit (solid)</li>
  <li><b>10% margin:</b> Bare 10 øre per krone (vanskelig å drive lønnsomt)</li>
  <li><b>Negativ margin:</b> Du taper penger på hver eneste salg</li>
</ul>

<p><b>Eksempel:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Net Profit: 458 kr
Sale Price: 799 kr

Margin % = (458 / 799) × 100 = 57.3%

Dette er VELDIG bra — over halvparten er profit.
</pre>

<p style="background:rgba(255,215,0,0.1); padding:10px; border-left:4px solid rgba(255,215,0,0.7); border-radius:3px">
  ⚠️ <b>Industry benchmark:</b> E-commerce trenger minimum 30-40% margin for å kunne drive profitabelt med paid ads.
</p>
<hr>

<h2 style="border-bottom:2px solid #555; padding-bottom:6px">🧩 Konsept 4 — CPA vs ROAS</h2>
<p>Dette er to sider av samme mynt — men de forteller deg <b>ulike ting</b>.</p>

<h3>📊 CPA (Cost Per Acquisition)</h3>
<p><b>Hva er det:</b> Hvor mye du betaler i ads for å få ETT salg.</p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
CPA = Total Ad Spend / Antall Salg
</pre>

<p><b>Eksempel:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Du brukte 5000 kr på ads
Fikk 25 salg

CPA = 5000 / 25 = 200 kr per salg
</pre>

<h3>📊 ROAS (Return On Ad Spend)</h3>
<p><b>Hva er det:</b> Hvor mye revenue du får for hver krone brukt på ads.</p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
ROAS = Total Revenue / Total Ad Spend
</pre>

<p><b>Eksempel:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Du brukte 5000 kr på ads
Fikk 20,000 kr i revenue

ROAS = 20,000 / 5000 = 4.0

Dette betyr: For hver krone brukt, fikk du 4 kr tilbake.
</pre>

<h3>🔗 Sammenhengen:</h3>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
ROAS = Sale Price / CPA

Hvis Sale Price = 799 kr og CPA = 200 kr:
ROAS = 799 / 200 = 3.995 ≈ 4.0
</pre>

<p style="background:rgba(100,149,237,0.1); padding:10px; border-left:4px solid rgba(100,149,237,0.5); border-radius:3px">
  💡 <b>Når bruke hva:</b><br>
  • Bruk <b>CPA</b> når du tenker kostnad ("Kan jeg tåle 200 kr per kunde?")<br>
  • Bruk <b>ROAS</b> når du tenker efficiency ("Får jeg nok revenue per krone brukt?")
</p>
<hr>

<h2 style="border-bottom:2px solid #555; padding-bottom:6px">🧩 Konsept 5 — Break-Even ROAS & Max CPA</h2>
<p>Dette er de <b>viktigste tallene</b> du må kjenne til.</p>

<h3>📊 Max CPA (Maximum Cost Per Acquisition)</h3>
<p><b>Hva er det:</b> Det høyeste du kan betale per salg uten å tape penger.</p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Max CPA = Net Profit (før ads)
</pre>

<p><b>Hvorfor?</b> Fordi hvis du bruker MER enn net profit på ads, taper du penger.</p>

<p><b>Eksempel:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Net Profit før ads: 458 kr

Max CPA = 458 kr

Hvis CPA > 458 kr → Du taper penger
Hvis CPA < 458 kr → Du tjener penger
Hvis CPA = 458 kr → Break-even (null profit)
</pre>

<h3>📊 Break-Even ROAS</h3>
<p><b>Hva er det:</b> Minimum ROAS du trenger for å gå i null.</p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Break-Even ROAS = Sale Price / Max CPA

Eller enklere:
Break-Even ROAS = 1 / Margin %
</pre>

<p><b>Eksempel:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
Sale Price: 799 kr
Max CPA: 458 kr

Break-Even ROAS = 799 / 458 = 1.74

Alternativt:
Margin % = 57.3% = 0.573
Break-Even ROAS = 1 / 0.573 = 1.74

Hvis ROAS > 1.74 → Profit
Hvis ROAS < 1.74 → Tap
</pre>

<p style="background:rgba(255,99,71,0.1); padding:10px; border-left:4px solid rgba(255,99,71,0.5); border-radius:3px">
  ⚠️ <b>KRITISK:</b> De fleste tror at ROAS > 2.0 alltid er bra. Men hvis din break-even ROAS er 2.5, taper du fortsatt penger ved 2.0 ROAS!
</p>
<hr>

<h2 style="border-bottom:2px solid #555; padding-bottom:6px">🧩 Konsept 6 — Margin Sensitivity</h2>
<p><b>Margin sensitivity</b> betyr: Små endringer i kostnader gir STORE endringer i lønnsomhet.</p>

<p><b>Eksempel — Fraktkostnad øker med 20 kr:</b></p>
<pre style="background:rgba(128,128,128,0.15); padding:10px; border-radius:5px; border:1px solid rgba(128,128,128,0.4); font-family:monospace">
OPPRINNELIG:
Sale Price: 799 kr
Costs: 341 kr (product + shipping + fees)
Net Profit: 458 kr
Margin: 57.3%
Break-Even ROAS: 1.74

ETTER FRAKTØKNING (+20 kr):
Sale Price: 799 kr (samme)
Costs: 361 kr (+20 kr)
Net Profit: 438 kr (-20 kr)
Margin: 54.8% (-2.5 prosentpoeng)
Break-Even ROAS: 1.82 (+0.08)
</pre>

<p><b>Hva skjedde?</b></p>
<ul>
  <li>Shipping økte bare 20 kr (5.9% av total cost)</li>
  <li>Men margin falt 2.5 prosentpoeng</li>
  <li>Break-even ROAS økte — nå trenger du BEDRE performance for å gå i null</li>
</ul>

<p style="background:rgba(255,215,0,0.1); padding:10px; border-left:4px solid rgba(255,215,0,0.7); border-radius:3px">
  ⚠️ <b>Lærdommen:</b> Produktutvikling handler ikke bare om å lage kule ting — det handler om å <b>beskytte marginen</b>. En liten kostnadøkning kan ødelegge lønnsomheten.
</p>

<h2 style="border-bottom:2px solid #555; padding-bottom:6px">📌 Implementering — Del 2: Test Break-Even Calculator</h2>

```python
# Produktscenario: Premium kaffetrakter
calc = BreakEvenCalculator(
    sale_price=799,
    product_cost=250,
    shipping_cost=65,
    payment_fee_percent=0.029,  # 2.9% (Stripe standard)
    payment_fee_fixed=3          # 3 kr fast fee
)

# Hent full analyse
analysis = calc.get_full_analysis()

# Print resultatene pent formatert
print("="*50)
print("📊 BREAK-EVEN ANALYSIS")
print("="*50)
print(f"Sale Price: {analysis['sale_price']} kr")
print(f"Total Costs: {analysis['total_costs']:.2f} kr")
print(f"Net Profit (før ads): {analysis['net_profit']:.2f} kr")
print(f"Margin: {analysis['margin_percent']:.1f}%")
print("-"*50)
print(f"Max CPA: {analysis['max_cpa']:.2f} kr")
print(f"Break-Even ROAS: {analysis['breakeven_roas']:.2f}")
print("="*50)
```

<h2 style="border-bottom:2px solid #555; padding-bottom:6px">📌 Visualisering — Margin Sensitivity Test</h2>
<p>La oss teste hva som skjer når produktkostnaden endrer seg.</p>

```python
# Test ulike produktkostnader
product_costs = np.arange(150, 450, 20)
margins = []
breakeven_roas_values = []

for cost in product_costs:
    temp_calc = BreakEvenCalculator(
        sale_price=799,
        product_cost=cost,
        shipping_cost=65,
        payment_fee_percent=0.029,
        payment_fee_fixed=3
    )
    result = temp_calc.get_full_analysis()
    margins.append(result['margin_percent'])
    breakeven_roas_values.append(result['breakeven_roas'])

# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Margin %
ax1.plot(product_costs, margins, linewidth=2, color='#2E86AB')
ax1.axhline(y=30, color='red', linestyle='--', label='Minimum viable (30%)')
ax1.set_xlabel('Product Cost (kr)', fontsize=12)
ax1.set_ylabel('Margin %', fontsize=12)
ax1.set_title('Hvordan produktkostnad påvirker margin', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Break-Even ROAS
ax2.plot(product_costs, breakeven_roas_values, linewidth=2, color='#A23B72')
ax2.axhline(y=2.0, color='orange', linestyle='--', label='"Standard" ROAS target (2.0)')
ax2.set_xlabel('Product Cost (kr)', fontsize=12)
ax2.set_ylabel('Break-Even ROAS', fontsize=12)
ax2.set_title('Hvordan produktkostnad påvirker break-even ROAS', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("💡 Innsikt: Jo høyere produktkostnad, desto høyere break-even ROAS trengs.")
```
